```python
from rdkit import Chem
import pandas as pd
import numpy as np
from seq_graph_retro.utils.parse import *
from canonicalize_prod import *
import rdkit.Chem.Draw as Draw
from rdkit import RDLogger
RDLogger.DisableLog('rdApp.*')

df = pd.read_csv("../datasets/mapped_rxn.csv")

id = []
for index in df['index']:
    id.append(index)

rxn = []
for smarts in df['mapped_rxn']: 
    rxn.append(smarts)

rxn_smi_map = list(zip(id, rxn))


for i,rxn in rxn_smi_map:
    r, p = rxn.split(">>")
    reac_mol = get_mol(r)
    prod_mol = get_mol(p)

    # img=Draw.MolsToGridImage([reac_mol, prod_mol], molsPerRow=2, subImgSize=(400,400))
    # display(img)
    if reac_mol is None or prod_mol is None:
        continue

    rxn_core, core_edits = get_reaction_core(r, p, kekulize=False, use_h_labels=True)
    print(i,':',rxn_core, core_edits)
```
